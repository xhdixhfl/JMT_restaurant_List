```python
########################## DR ############################
# 별점평균 구하는 함수

# RT_List = []
# RV_List = []
# List = []
def set_RT_List():
    sql_RT = 'select * from RT_List'
    cursor.execute(sql_RT)
    # for 개별변수 in 집합변수
    for row in cursor:
        RT_List.append(row)
    return RT_List

def set_RV_List():
    sql_RV = 'select * from RV_List'
    cursor.execute(sql_RV)
    # for 개별변수 in 집합변수
    for row in cursor:
        RV_List.append(row)
    return RV_List

# 가게이름으로 평균점수 구하기
def avg_RV_List(a):
    RV_L = set_RV_List()
    List = []
    for i in range(len(RV_L)):
        
        if RV_L[i][0]== a:
            List.append(RV_L[i])
    tot = 0
    for j in List:
        tot += j[4]
    avg = tot/len(List)
            
    return avg
```
